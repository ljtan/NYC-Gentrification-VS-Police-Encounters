# This notebook is used to process the ACS data saved as pop_10.csv and pop_18.csv

https://www.baruch.cuny.edu/confluence/display/geoportal/NYC+Geodatabase

```python
import pandas as pd
import numpy as np
```

```python
lookup = pd.read_excel('data/nyc_zcta_census_data.xlsx', 'b_2018acs_lookup')
```

```python
# find column name for total population
lookup[lookup['est_value'].str.contains('Total population')]
```

```python
pop_18_1 = pd.read_excel('data/nyc_zcta_census_data.xlsx', 'b_zctas_2018acs1')
pop_18_1 = pop_18_1[['GEOID2', 'EDU10_PC', 'EMP05_PC', 'INC12_E']]
pop_18_1.columns = ['zip', 'college_pct', 'unemployment_pct', 'median_income']
pop_18_1
```

```python
# open data 2018
pop_18_2 = pd.read_excel('data/nyc_zcta_census_data.xlsx', 'b_zctas_2018acs2')
pop_ls_2=['GEOID2','SXAG01_E','HOC01_E', 'HOC02_PC','HTEN02_PC', 'HVAL01_E', 
          'RENT09_E', 'RACE03_PC', 'RACE04_PC', 'RACE06_PC', 'HISL02_PC']
pop_18_2 = pop_18_2[pop_ls_2]
pop_18_2.columns = ['zip', 'population', 'housing_units', 'occu_pct','own_occu_pct', 'median_home_val', 'median_rent','white_pct','black_pct','asian_pct','hisp_pct']
pop_18_2
```

```python
pop_18 = pop_18_2.merge(pop_18_1, left_on='zip', right_on='zip')
pop_18
```

```python
pop_18.columns
```

```python
pop_10_1 = pd.read_excel('data/nyc_zcta_census_data.xlsx', 'b_zctas_2010census')
pop_10_1 = pop_10_1[['GEOID2', 'HD01_S001', 'HD01_S169']]
pop_10_1.columns = ['zip', 'population', 'housing_units']
pop_10_1
```

```python
pop_10_2 = pd.read_excel('data/nyc_zcta_census_data.xlsx', 'b_zctas_2010census_pct')
pop_10_2 = pop_10_2[['GEOID2','HD02_S170','HD02_S181','HD02_S078', 'HD02_S079', 'HD02_S081', 'HD02_S114']]
pop_10_2.columns = ['zip','occu_pct', 'own_occu_pct', 'white_pct', 'black_pct', 'asian_pct', 'hisp_pct']
pop_10_2
```

```python
pop_10 = pop_10_1.merge(pop_10_2, left_on='zip', right_on='zip')
pop_10
```

```python
pop_10.to_csv('data/pop_10.csv')
pop_18.to_csv('data/pop_18.csv')
```
